# Named entity recognition

```python
import nltk
from nltk import word_tokenize, pos_tag, punkt, ne_chunk
# Named entity chunking of POS tagged tokens, and recognition of the same

test_description = "Withdrawal from ATM ABC1234 on 05/01/2022"
tokens = word_tokenize(test_description)
tagged_tokens = pos_tag(tokens)
ne_tree = ne_chunk(tagged_tokens)
named_entities = []
for subtree in ne_tree.subtrees():
    if subtree.label() in ['ORGANIZATION', 'PERSON', 'LOCATION']:
        named_entity = " ".join([token for token, pos in subtree.leaves()])
        named_entities.append(named_entity)
print(named_entities)
```
